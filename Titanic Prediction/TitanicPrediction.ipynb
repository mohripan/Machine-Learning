```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
train = pd.read_csv('https://raw.githubusercontent.com/mohripan/Kaggle-Training/main/Titanic%20Prediction/train.csv')
test = pd.read_csv('https://raw.githubusercontent.com/mohripan/Kaggle-Training/main/Titanic%20Prediction/test.csv')
train.head()
```

```python
train.info()
```

```python
X = train.drop('Survived', axis=1)
y = train['Survived']
```

```python
data = pd.concat([X, test])
data
```

```python
data.shape
```

```python
data.info()
```

```python
data['Cabin'].fillna('missing', inplace=True)
```

```python
kosong = []
for i in data['Cabin']:
  if i != 'missing':
    kosong.append(i[0])
  else:
    kosong.append('missing')

data['FirstCabin'] = kosong
```

```python
data['FirstCabin'].unique()
```

```python
non_numbers = []
for i in data['Ticket']:
  result = ''.join([j for j in i if not j.isdigit()])
  if result != '':
    non_numbers.append(result.replace('.','').replace('/',''))
  else:
    non_numbers.append('contain_number')

data['CodeTicket'] = non_numbers
```

```python
numbers = []
for i in data['Ticket']:
  result = ''.join([j for j in i if j.isdigit()])
  if result != '':
    numbers.append(result)
  else:
    numbers.append('not_contain_number')

data['NumberTickets'] = numbers
```

```python
kosong = []
is_missing = []
for i in data['Cabin']:
  if i != 'missing':
    kosong.append(i[1:])
    is_missing.append(False)
  else:
    kosong.append(0)
    is_missing.append(True)

data['SecondCabin'] = kosong
```

```python
data = data.drop('PassengerId', axis=1)
```

```python
data = data.drop('Name', axis=1)
data = data.drop('Cabin', axis=1)
data = data.drop('Ticket', axis=1)
```

```python
numeric = ['int64', 'float64']

for i in data.columns:
  if data[i].dtype in numeric:
    data[i].fillna(data[i].median(), inplace=True)

data.info()
```

```python
data['Embarked'].fillna('missing', inplace=True)
```

```python
data = pd.get_dummies(data)
```

```python
X = data.iloc[:len(train), :]
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from xgboost import XGBRFClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LogisticRegressionCV
from lightgbm import LGBMClassifier

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
#estimators = [('rf', make_pipeline(StandardScaler(), LGBMClassifier(n_estimators=100))),('xgbc', make_pipeline(StandardScaler(), XGBClassifier(n_estimators=100)))]

model = make_pipeline(StandardScaler(), XGBClassifier(n_estimators=1000, max_depth=3, learning_rate=0.01))
model.fit(X_train, y_train)
print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
```

```python
testing = data.iloc[len(train):, :]
y_pred = model.predict(testing)
submission = pd.DataFrame({'PassengerId': test['PassengerId'], 'Survived': y_pred})
submission.to_csv('submit.csv', index=False)
```
